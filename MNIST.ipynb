```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
mnist=pd.read_csv("/content/sample_data/mnist_train_small.csv")
```

```
mnist.head()
```

```
mnist.info()
mnist.columns
```

```
mnist.describe()
```

```
from sklearn.preprocessing import StandardScaler
st = StandardScaler()
st.fit(mnist.drop('6',axis=1))
mnist_train=st.transform(mnist.drop('6',axis=1))
```

```
from sklearn.decomposition import PCA
p = PCA(n_components=173)
out = p.fit_transform(mnist_train)
```

```
principalDf = pd.DataFrame(data = out)
```

```
fin = pd.concat([principalDf,mnist[['6']]],axis=1)
```

```
fin.head()
```

```
from sklearn.svm import SVC
sv = SVC()
sv.fit(fin.drop('6',axis=1),fin['6'])
```

```
mnist_test=pd.read_csv('/content/sample_data/mnist_test.csv')
```

```
mnist_test.info()
```

```
mntest=st.transform(mnist_test.drop('7',axis=1))
```

```
from sklearn.decomposition import PCA
p = PCA(n_components=173)
out = p.fit_transform(mntest)
```

```
principalDf1 = pd.DataFrame(data = out)
```

```
fin1 = pd.concat([principalDf1,mnist_test[['7']]],axis=1)
```

```
re=sv.predict(fin1.drop('7',axis=1))
```

```
print(re)
```
